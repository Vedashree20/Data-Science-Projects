```python
# let's have a look at how many labels
for col in df.columns[0:]:
    print(col, ': ', len(df[col].unique()), ' labels')
```

```python
pd.get_dummies(df,drop_first=True).shape
```

```python
# let's find the top 10 most frequent categories for the variable Version

df.Version.value_counts().sort_values(ascending=False).head(8)
```

```python
# let's make a list with the most frequent categories of the variable

top_5_labels = [x for x in df.Version.value_counts().sort_values(ascending=False).head(5).index]
top_5_labels
```

```python
import numpy as np
for label in top_5_labels:
     df[label] = np.where(df['Version']==label, 1, 0)
df[['Version']+top_5_labels].head(5)
```

```python
def one_hot_encoding_top_x(df, variable, top_x_labels):
    # function to create the dummy variables for the most frequent labels
    # we can vary the number of most frequent labels that we encode
    
    for label in top_x_labels:
        df[variable+'_'+label] = np.where(df[variable]==label, 1, 0)
# read the data again
df2 = df[['Brand_Name', 'Version']]

# encode X2 into the 10 most frequent categories
one_hot_encoding_top_x(df2, 'Version', top_5_labels)
df2.head()
```
